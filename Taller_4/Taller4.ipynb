## Laplaciano

```python
spot_imagen = cv2.imread('images/SPOT.png', cv2.IMREAD_GRAYSCALE)
spot_imagen_i16 = spot_imagen.astype(np.int16)

plt.imshow(spot_imagen, cmap='gray')
plt.axis('off')
plt.show()
```

### 3.4 Aplique un filtro Laplaciano. Visualice el resultado ****(puede efectuar una calibración del histograma -entre 0 y 255- de la imagen obtenida para visualizar mejor los detalles de la imagen del Laplaciano).

```python
from skimage.filters import laplace
from skimage import exposure

spot_laplaciano = laplace(spot_imagen_i16, ksize=3)  

plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.title("Original")
plt.imshow(spot_imagen, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.subplot(1,3,2)
plt.title("Laplaciano")
plt.imshow(spot_laplaciano, cmap='gray') 
plt.axis('off')
plt.show()
```

### 3.5 ¿Cómo es la respuesta del Laplaciano en las regiones homogéneas de la imagen y en los contornos? ¿A qué se debe este resultado?

Podemos ver que en las regiones homogéneas la respuesta del Laplaciano es cercana a cero, porque al ser una medida de la segunda derivada espacial no hay variación de intensidad y, por tanto, la curvatura de la función imagen es prácticamente nula. Por otro lado, en los contornos el Laplaciano toma valores de gran magnitud y cambia de signo a ambos lados del borde, ya que allí la intensidad cambia rápidamente y la segunda derivada se vuelve significativa.

### 3.6 Aplique sobre el resultado del paso anterior un filtro Laplaciano. Visualice el resultado ****(puede efectuar una calibración del histograma -entre 0 y 255- de la imagen obtenida para visualizar mejor los detalles de la imagen del Laplaciano). ¿Cómo es la respuesta de este LoG con respecto al Laplaciano sin filtrado previo del ejercicio anterior? ¿A qué se debe este resultado?

```python
from skimage.filters import gaussian

spot_gaussian = gaussian(spot_imagen, sigma=0, preserve_range=True)

plt.imshow(spot_gaussian, cmap='gray')
plt.axis('off')
plt.show()
```

```python
spot_laplaciano = laplace(spot_gaussian, ksize=3)  

plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.title("Original (σ = 0)")
plt.imshow(spot_gaussian, cmap='gray')
plt.axis('off')
plt.subplot(1,3,2)
plt.title("Laplaciano (σ = 0)")
plt.imshow(spot_laplaciano, cmap='gray') 
plt.axis('off')
plt.show()
```

Podemos observar que, en este caso, la respuesta no se diferencia del ejercicio anterior. Esto se debe a que usamos una varianza igual a 0 en el filtrado gaussiano, lo que equivale prácticamente a no aplicar suavizado. Por esta razón, el operador LoG con una gaussiana tan angosta se comporta como el Laplaciano aplicado directamente sobre la imagen.

### 3.7 Repita los pasos 1 y 2, pero cambiando a 1 el valor de la varianza en el filtro gaussiano. ¿Cómo es la respuesta de este LoG con respecto al Laplaciano sin filtrado previo del ejercicio anterior? ¿A qué se debe este resultado?

```python
from skimage.filters import gaussian

spot_gaussian = gaussian(spot_imagen, sigma=1, preserve_range=True)

plt.imshow(spot_gaussian, cmap='gray')
plt.axis('off')
plt.show()
```

```python
spot_laplaciano = laplace(spot_gaussian, ksize=3)  

plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.title("Original (σ = 1)")
plt.imshow(spot_gaussian, cmap='gray')
plt.axis('off')
plt.subplot(1,3,2)
plt.title("Laplaciano (σ = 1)")
plt.imshow(spot_laplaciano, cmap='gray') 
plt.axis('off')
plt.show()
```

Con una varianza igual a 1, la respuesta del LoG es similar a la del Laplaciano pero se observa más suave y con menos variaciones pequeñas o artefactos, por esto los bordes se ven menos intensos pero más limpios en comparación a la anterior. Esto se debe a que el filtrado gaussiano previo atenúa las componentes de alta frecuencia antes de aplicar la segunda derivada.

### 3.8 Repita los pasos 1 y 2, pero cambiando a 2 el valor de la varianza en el filtro gaussiano. ¿Cómo es la respuesta de este LoG? ¿A qué se debe este resultado?

```python
spot_gaussian = gaussian(spot_imagen, sigma=2, preserve_range=True)

plt.imshow(spot_gaussian, cmap='gray')
plt.axis('off')
plt.show()
```

```python
spot_laplaciano = laplace(spot_gaussian, ksize=3)  

plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.title("Original (σ = 2)")
plt.imshow(spot_gaussian, cmap='gray')
plt.axis('off')
plt.subplot(1,3,2)
plt.title("Laplaciano (σ = 2)")
plt.imshow(spot_laplaciano, cmap='gray') 
plt.axis('off')
plt.show()
```

Con una varianza igual a 2 el LoG se ve todavía más suavizado, lo que reduce aún más los detalles finos y hace que solo permanezcan las estructuras más grandes. Por ello, los contornos aparecen más gruesos y el realce es menos pronunciado. Esto se debe a que, al aumentar la varianza, el filtro gaussiano promedia sobre una vecindad mayor y suprime con mayor fuerza las altas frecuencias antes de aplicar el Laplaciano.

### 3.9 Cargue la imagen bruit y visualícela. Aplique un filtro gaussiano (con varianza igual a 1) y un filtro Laplaciano siguiendo los mismos pasos de los ejercicios previos. Visualice la imagen del LoG.

```python
bruit_imagen = cv2.imread('images/BRUIT.png', cv2.IMREAD_GRAYSCALE)

plt.imshow(bruit_imagen, cmap='gray')
plt.axis('off')
plt.show()
```

```python
bruit_gaussian = gaussian(bruit_imagen, sigma=1, preserve_range=True)

plt.imshow(bruit_gaussian, cmap='gray')
plt.axis('off')
plt.show()
```

```python
bruit_laplaciano = laplace(bruit_gaussian, ksize=3)  

plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.title("Original (σ = 1)")
plt.imshow(bruit_gaussian, cmap='gray')
plt.axis('off')
plt.subplot(1,3,2)
plt.title("Laplaciano (σ = 1)")
bruit_laplaciano = laplace(bruit_gaussian, ksize=3)  
plt.imshow(bruit_laplaciano, cmap='gray') 
plt.axis('off')
plt.show()
```

### 3.10 Calcule sobre la misma imagen bruit la norma del gradiente utilizando el operador de Sobel. Para una mejor visualización, calibre el histograma de las imágenes resultantes. ****

```python
from skimage.filters import sobel
from skimage.exposure import rescale_intensity

bruit_gradiente = sobel(bruit_imagen)
bruit_calibrado = rescale_intensity(bruit_gradiente, in_range='image', out_range=(0, 255)).astype(np.uint8)  # [web:51]

plt.figure(figsize=(10,4))
plt.subplot(1,2,1); plt.title("Norma del gradiente")
plt.imshow(bruit_gradiente, cmap='gray') 
plt.axis('off')
plt.subplot(1,2,2); plt.title("Norma del gradiente calibrado ")
plt.imshow(bruit_calibrado, cmap='gray', vmin=0, vmax=255) 
plt.axis('off')
plt.show()
```

### 3.11 ¿En cuál de las dos imágenes es más fácil detectar los contornos. ¿Por qué ? ¿Cómo afecta el ruido la detección de los contornos en cada caso?

Es más fácil detectar los contornos en la imagen obtenida con Sobel, ya que la norma del gradiente produce una respuesta positiva y marcada en los bordes que el Laplaciano. Con respecto al ruido, este afecta a ambos métodos, pero suele afectar más al Laplaciano porque amplifica con mayor intensidad las altas frecuencias. Por eso, el LoG mejora con respecto al Laplaciano directo, ya que el suavizado gaussiano previo reduce el ruido antes de aplicar el Laplaciano.

```python
film_imagen = cv2.imread('images/FILM.png', cv2.IMREAD_GRAYSCALE)

plt.imshow(film_imagen, cmap='gray')
plt.axis('off')
plt.show()
```

### 3.13 Aplique un filtro gaussiano (con varianza igual a 1) y un filtro Laplaciano siguiendo los mismos pasos de los ejercicios previos. Visualice la imagen del LoG.

```python
from skimage.filters import gaussian

film_gaussian = gaussian(film_imagen, sigma=1, preserve_range=True)

plt.imshow(film_gaussian, cmap='gray')
plt.axis('off')
plt.show()
```

```python
film_laplaciano = laplace(film_gaussian, ksize=3)  

plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.title("Original (σ = 1)")
plt.imshow(film_gaussian, cmap='gray')
plt.axis('off')
plt.subplot(1,3,2)
plt.title("Laplaciano (σ = 1)")
film_laplaciano = laplace(film_gaussian, ksize=3)  
plt.imshow(film_laplaciano, cmap='gray') 
plt.axis('off')
plt.show()
```

### 3.14 Calibre el histograma de la imagen del LoG entre 0 y 255 y déjela de tipo UInt8.

```python
film_calibrado = rescale_intensity(film_laplaciano, in_range='image', out_range=(0, 255)).astype(np.uint8) 

plt.figure(figsize=(10,4))
plt.subplot(1,2,2); plt.title("LoG calibrado ")
plt.imshow(film_calibrado, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()
```

### 3.15 Combine esta última imagen (la del LoG calibrada) con 60% de la imagen inicial. Esto quiere decir multiplicar la imagen original por el valor 0.6 y sumarla a la imagen invertida del LoG. 

```python
invertida = 255 - film_calibrado  
film_f = film_imagen.astype(np.float32)
film_invertida = invertida.astype(np.float32)
combinada = 0.6 * film_f + film_invertida
combinada_u8 = np.clip(combinada, 0, 255).astype(np.uint8)  

plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.title("Original")
plt.imshow(film_imagen, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.subplot(1,3,2); plt.title("LoG invertido")
plt.imshow(invertida, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.subplot(1,3,3); plt.title("Combinada")
plt.imshow(combinada_u8, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()
```

### 3.16 Para acentuar el interés del procesamiento, efectúe una ecualización del histograma de la imagen original y de la imagen combinada. Visualice el resultado. ¿Cuál es su conclusión?

```python
film_ecualizada = cv2.equalizeHist(film_imagen)      
combinada_ecualizada = cv2.equalizeHist(combinada_u8)      

plt.figure(figsize=(12,4))
plt.subplot(1,2,1); plt.title("Original ecualizada")
plt.imshow(film_ecualizada, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.subplot(1,2,2); plt.title("Combinada ecualizada")
plt.imshow(combinada_ecualizada, cmap='gray', vmin=0, vmax=255) 
plt.axis('off')
plt.show()
```

La ecualización del histograma mejora el contraste global al redistribuir los niveles de gris para que ocupen mejor todo el rango, haciendo más visibles detalles que antes se veían apagados tanto en la imagen original como en la combinada. En la imagen combinada, el efecto de la ecualización potencia aún más el realce de bordes y texturas, pero también hace más notorio el ruido en las zonas homogéneas.

## Ejercicio de síntesis taller 4

```python
amour_imagen = cv2.imread('images/AMOUR.png', cv2.IMREAD_GRAYSCALE)

plt.imshow(amour_imagen, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()
```

```python
s = sobel(amour_imagen)
_, edges = cv2.threshold(s, s.max()/3, 255, cv2.THRESH_BINARY)  
out = amour_imagen.copy()
out[edges > 0] = 180

plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.title("Original"); plt.imshow(amour_imagen, cmap='gray', vmin=0, vmax=255) 
plt.axis('off')
plt.subplot(1,3,2); plt.title("Bordes"); plt.imshow(edges, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.subplot(1,3,3); plt.title("Resultado"); plt.imshow(out, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()
```
